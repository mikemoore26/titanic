```python
def reduce_name(i):
    i = i.split(",")[1].split(".")[0]
    if i.strip() == "Miss":
         i = "Ms"
    else:
        i = i.strip()
    #print(i)
    return i
```

```python
def fix_data(data):
    #print(data.head(1))
    drop_cols = ["Sex","Name","PassengerId","Ticket","Cabin","Embarked"]
    
    #make new columns
    data['Is_Male'] = data['Sex'].map(lambda x: 1 if x == "male" else 0)
    data['Pre_Name'] = data.Name.apply(lambda x: reduce_name(x))
    data['Age_Cat'] = pd.cut(data['Age'],bins=10, labels=False)
    #remove columns
    data.drop(drop_cols,1,inplace=True)
    
    data = pd.get_dummies(data)
    
    data = data.dropna()

    return data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Survived",1)
y = df["Survived"]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.25,random_state=26)
```

```python
print(f'The Dim of xtrain is {xtrain.shape}')
print(f'The Dim of xtest is {xtest.shape}')
print(f'The Dim of ytrain is {ytrain.shape}')
print(f'The Dim of ytest is {ytest.shape}')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X,y)
pred_y = model.predict(xtest)
pred_y

res_df=pd.DataFrame()
res_df['True'] = ytest
res_df['Predicted'] = pred_y
res_df
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
models =[XGBClassifier(),LinearDiscriminantAnalysis(),KNeighborsClassifier(3),GaussianNB(),DecisionTreeClassifier(),SVC(),LogisticRegression()]
```

```python
def analyse_model(model,tx,ty):
    print("*"*20)
    pred = model.predict(tx)
#     print(classification_report(ty,pred))
#     print('Mean Accuracy: %.3f' % model.best_score_)
#     print('Config: %s' % model.best_params_)
#     print("*"*20)
    sns.heatmap(confusion_matrix(ytest,pred), annot=True, cbar=False)
    plt.ylabel("True")
    plt.xlabel("Predicted")
    plt.show()

    print("*"*20)
```

```python
def compare_Models(x,y,tx,ty,models):
    acc_score = {}
    best_acc = 0
    for model in models:
        model.fit(x,y)
        p_y = model.predict(tx)
        acc = accuracy_score(ty,p_y)
        if acc > best_acc:
            best_acc = acc
        acc_score[model] = acc
        print(model)
        analyse_model(model,xtest,ytest)

        #print(f"accuracy score of the {model} model is {round(acc*100,4)}")
    for i in acc_score.items():
        if i[1] == best_acc:
            print(f"{i[0]}-->{i[1]} *BEST*")
            bestm = i
        else:
            print(f"{i[0]}-->{i[1]}")
            
        
    return bestm[0]
    
    
```

```python
model = compare_Models(xtrain,ytrain,xtest,ytest,models)
```

```python
analyse_model(model,xtest,ytest)
```

```python
from sklearn.model_selection import GridSearchCV, RepeatedStratifiedKFold
import numpy as np
```

```python
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
```

```python
grid = dict()
grid['solver'] = ['svd', 'lsqr', 'eigen']
grid['shrinkage'] = np.arange(0, 1, 0.01)
# define search
search = GridSearchCV(model, grid, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
results = search.fit(xtest, ytest)
```

```python
analyse_model(search,xtest,ytest)
```
